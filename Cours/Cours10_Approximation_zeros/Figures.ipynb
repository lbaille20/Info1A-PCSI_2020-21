<h2 style='text-align:center'>Approximation des zéros d'une fonction (TP10) - v1</h2>

* **Cadre général :**   
    On cherche à approcher par des flottants la valeur du zéro $\alpha$ d'une fonction $f$ sur un segment (intervalle fermé borné) $\left[a, b\right]$ de $\mathbb R$ ($a<b$), et vérifiant les conditions suivantes :  
    1. $f$ est continue sur $\left[a, b\right]$ ;  
    2. $f$ est strictement monotone sur $\left[a, b\right]$ ;  
    3. $f$ change de signe sur $\left[a, b\right]$ (ce que l'on caractérise par le fait que $f(a)f(b)<0$).  
  
Ces trois conditions assurent que l'équation $f(x)=0$ admet une unique solution sur $\left[a, b\right]$ (corollaire du théorème des valeurs intermédiaires).  

*Note :* on appelle **zéro** de $f$ sur $\left[a, b\right]$, tout réel $x$ tel que $x\in \left[a, b\right]$ et $f(x)=0$.

* Exemple :

La fonction $f : x \longmapsto x^2 - 2$, vérifie, sur le segment $\left[a, b\right]=\left[1, 2\right]$ les trois conditions précédentes, et donc admet sur le segment $\left[a, b\right]=\left[1, 2\right]$ un unique **zéro**.  
Cet unique zéro est bien évidemment le réel $\sqrt(2)$.

&#9788; illustration graphique :

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: x**2 - 2

xmin, xmax, ymin, ymax = -1, 3, -2, 2

subd = np.linspace(xmin, xmax, 100 + 1)
ords = f(subd)

## courbe représentative de f
plt.plot(subd, ords, linestyle = '-', color = 'blue', label = 'Courbe représentative de $f$')

## zéro de la fonction
plt.plot([np.sqrt(2)], [0], color = 'orange', marker = 'o')

## annotation du zéro
## https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html
## https://matplotlib.org/stable/api/text_api.html#matplotlib.text.Text
## https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.ArrowStyle.html
plt.annotate(r'$\sqrt{2}$', xy=(np.sqrt(2) + 0.05, 0  + 0.05), xytext=(1.75, 0.5), color = 'orange',
             arrowprops=dict(arrowstyle = '->', color='orange'))

## axes des abscisses et des ordonnées 
plt.plot([xmin, xmax], [0, 0], 'k-')
plt.plot([0, 0], [ymin, ymax], 'k-')

## fenêtre de visualisation
plt.axis([xmin, xmax, ymin, ymax])

## affichage de la légende
plt.legend()

## affichage
plt.show()
```

## I) Méthode de dichotomie

Les conditions de l'introduction sont supposées vérifiées par une fonction $f$ sur un segment $\left[a, b\right]$.  
L'unique zéro de $f$ sur $\left[a, b\right]$ est noté $\alpha$.    
On définit deux suites $\left(a_n\right)_{n\geqslant 0}$ et $\left(b_n\right)_{n\geqslant 0}$ par :  
* $a_0=a$ et $b_0=b$ ;  
* pour tout $n\geqslant 0$ : $\quad$ on note $c_n=\frac{a_n+b_n}{2}$, et 
    - si $f(c_n)$ a la même signe que $f(a)$, *i.e.* si 
    $f(a).f(c_n)\geqslant 0$
    : $\quad$ $a_{n+1} = c_n =\frac{a_n+b_n}{2}$
    et $b_{n+1} = b_n$ ;
    - sinon, *i.e.* si 
    $f(a).f(c_n)< 0$
    : $\quad$ $a_{n+1} = a_n$
    et $b_{n+1} = c_n =\frac{a_n+b_n}{2}$.

```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: x**2 - 2

xmin, xmax, ymin, ymax = -1, 3, -2, 2

subd = np.linspace(xmin, xmax, 100 + 1)
ords = f(subd)

## dimensionnement de la fenêtre
## https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html
plt.figure(figsize = (20, 10))
## taille de la police
## https://stackoverflow.com/questions/3899980/how-to-change-the-font-size-on-a-matplotlib-plot
plt.rcParams.update({'font.size': 22})

## courbe représentative de f
plt.plot(subd, ords, linestyle = '-', color = 'blue', label = 'Courbe représentative de $f$')

## zéro de la fonction
plt.plot([np.sqrt(2)], [0], color = 'orange', marker = 'o', markersize = 10)

a, b = 1, 2

a0, b0 = a, b
plt.plot([a0], [0], 'ro', markersize = 10)
plt.plot([b0], [0], 'ro', markersize = 10)

c = (a0 + b0) / 2
plt.plot([c], [0], color = 'violet', marker = 'o', markersize = 10)

## flèches


## annotations
plt.annotate(r'$a_0$', xy=(a0 - 0.05, 0  + 0.05), xytext=(a0 - 0.5, 0.5), color = 'red',
             arrowprops=dict(arrowstyle = '->', color='red'))
plt.annotate(r'$b_0$', xy=(b0 + 0.05, 0  + 0.05), xytext=(b0 + 0.5, 0.5), color = 'red',
             arrowprops=dict(arrowstyle = '->', color='red'))
plt.annotate(r'$c_0=\frac{a_0+b_0}{2}$', xy=(c + 0.05, 0  + 0.05), xytext=(c + 0.5, 1.0), color = 'violet',
             arrowprops=dict(arrowstyle = '->', color='violet'))
plt.annotate(r'$\alpha$', xy=(np.sqrt(2) - 0.05, 0  + 0.05), xytext=(np.sqrt(2) - 0.5, 1.5), color = 'orange',
             arrowprops=dict(arrowstyle = '->', color='orange'))

## axes des abscisses et des ordonnées 
plt.plot([xmin, xmax], [0, 0], 'k-')
plt.plot([0, 0], [ymin, ymax], 'k-')

## fenêtre de visualisation
plt.axis([xmin, xmax, ymin, ymax])

## affichage de la légende
plt.legend()

## affichage
plt.show()
```

Propriétés des suites $\left(a_n\right)_{n\geqslant 0}$ et $\left(b_n\right)_{n\geqslant 0}$

* pour tout $n\geqslant 0$ : $$a\leqslant a_n\leqslant b_n\leqslant b \quad \text{   et   }\quad
f(a_n)\times f(b_n)\leqslant 0.$$

* pour tout $n\geqslant 0$ : $f$ continue, strictement monotone sur $[a_n,b_n]$

et donc pour tout $n\geqslant 0$ : $a_n\leqslant \alpha\leqslant b_n$

On peut montrer encore que les suites $\left(a_n\right)_{n\geqslant 0}$ et $\left(b_n\right)_{n\geqslant 0}$ sont **adjacentes** :
* $\left(a_n\right)_{n\geqslant 0}$ croissante  
* $\left(b_n\right)_{n\geqslant 0}$ décroissante  

et 
$$ 0\leqslant b_n-a_n = \frac{b-a}{2^n}$$

```python
## animation
%matplotlib notebook
## https://stackoverflow.com/questions/4098131/how-to-update-a-plot-in-matplotlib

import matplotlib.pyplot as plt
import  numpy as np

fig = plt.figure(figsize = (12, 8))
plt.rcParams.update({'font.size': 22})
## axes des abscisses et des ordonnées 
plt.plot([xmin, xmax], [0, 0], 'k-')
plt.plot([0, 0], [ymin, ymax], 'k-')

## fenêtre de visualisation
plt.axis([xmin, xmax, ymin, ymax])
plt.ion()

f = lambda x: x**2 - 2

xmin, xmax, ymin, ymax = -1, 3, -2, 2
xmin, xmax, ymin, ymax = 0.5, 2., -2, 2

subd = np.linspace(xmin, xmax, 100 + 1)
ords = f(subd)

## courbe représentative de f
plt.plot(subd, ords, linestyle = '-', color = 'blue', label = 'Courbe représentative de $f$')

a, b = 1., 2.
print(a,b)
p, = plt.plot([a,b],[0.,0.],'ro-')
for k in range(8):
    m = (a+b)/2
    if f(a)*f(m)<0:
        b = m
    else:
        a = m
    print(a,b)
    plt.pause(1)
    p.set_xdata([a,b])
    fig.canvas.draw()
    fig.canvas.flush_events()
plt.show()
```

## II) Méthode de Newton

Les conditions de l'introduction sont supposées vérifiées par une fonction $f$ sur un segment $\left[a, b\right]$.  
L'unique zéro de $f$ sur $\left[a, b\right]$ est noté $\alpha$.    
On définit **une** suite $\left(c_n\right)_{n\geqslant 0}$ par :  
* la donnée d'un réel $c_0\in \left[a, b\right]$ ;  
* pour tout $n\geqslant 0$ : 
$$c_{n+1} \; \text{est l'abscisse du point d'intersection de la tangente à la courbe représentative de $f$ avec l'axe des abscisses}$$ 

```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: x**2 - 2

xmin, xmax, ymin, ymax = -1, 3, -2, 2

subd = np.linspace(xmin, xmax, 100 + 1)
ords = f(subd)

## dimensionnement de la fenêtre
## https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html
plt.figure(figsize = (20, 10))
## taille de la police
## https://stackoverflow.com/questions/3899980/how-to-change-the-font-size-on-a-matplotlib-plot
plt.rcParams.update({'font.size': 22})

## courbe représentative de f
plt.plot(subd, ords, linestyle = '-', color = 'blue', label = 'Courbe représentative de $f$')

## zéro de la fonction
plt.plot([np.sqrt(2)], [0], color = 'orange', marker = 'o', markersize = 10)

c = 1

c0 = c
plt.plot([c0], [0], 'ro', markersize = 10)
c0 = c
plt.plot([c0], [f(c0)], 'ro', markersize = 10)
plt.plot([c0, c0], [0, f(c0)], 'r--')

c1 = - f(c0) / (2 * c0) + c0
plt.plot([c1], [0], color = 'violet', marker = 'o', markersize = 10)
plt.plot([xmin, xmax], [2*c0*(xmin - c0) + f(c0), 2*c0*(xmax - c0) + f(c0)], 'r-.',
        label = 'tangente à $C_f$ au point d\'abscisse $c_0$')

## annotations
plt.annotate(r'$c_0$', xy=(c0 - 0.05, 0  + 0.05), xytext=(c0 - 0.5, 0.5), color = 'red',
             arrowprops=dict(arrowstyle = '->', color='red'))

plt.annotate(r'$c_1$', xy=(c1 + 0.05, 0  + 0.05), xytext=(c1 + 1.0, 1.0), color = 'violet',
             arrowprops=dict(arrowstyle = '->', color='violet'))

plt.annotate(r'$\alpha$', xy=(np.sqrt(2) - 0.05, 0  + 0.05), xytext=(np.sqrt(2) - 0.5, 1.5), color = 'orange',
             arrowprops=dict(arrowstyle = '->', color='orange'))

## axes des abscisses et des ordonnées 
plt.plot([xmin, xmax], [0, 0], 'k-')
plt.plot([0, 0], [ymin, ymax], 'k-')

## fenêtre de visualisation
plt.axis([xmin, xmax, ymin, ymax])

## affichage de la légende
plt.legend()

## affichage
plt.show()
```
